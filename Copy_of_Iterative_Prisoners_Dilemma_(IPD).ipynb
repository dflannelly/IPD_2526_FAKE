# Making Functions Smarter.
Both the cooperate always function and the wild card function did not take advantage of looking at past rounds to see what their opponent did. Let's create a smarter function that can play nice with friendly functions but carries a grudge against other functions.

```
#Don't Cross Me
def Dont_Cross_Me():
  #for the first round cooperate
  if len(my_history)==0:
    return('c')
  elif 'b' in their_history:
    return('b')
  else:
    return('c')
```
